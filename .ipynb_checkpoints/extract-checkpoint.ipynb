```python
# scanner_easyocr.py
import easyocr
from PIL import Image
import tkinter as tk
from tkinter import filedialog, messagebox
import os

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'])  # English

def scan_image():
    file_path = filedialog.askopenfilename(
        title="Select an Image",
        filetypes=[("Image Files", "*.png *.jpg *.jpeg *.bmp *.tiff")]
    )
    if not file_path:
        return
    
    try:
        image = Image.open(file_path)
        image.show()  # Optional
        
        # OCR using EasyOCR
        result = reader.readtext(file_path)
        
        # Combine extracted text
        extracted_text = "\n".join([res[1] for res in result])
        
        # Display in new window
        result_window = tk.Toplevel(root)
        result_window.title("Extracted Text")
        result_window.geometry("600x400")
        
        text_box = tk.Text(result_window, wrap=tk.WORD)
        text_box.insert(tk.END, extracted_text)
        text_box.pack(expand=True, fill=tk.BOTH)
        
        # Save text file
        save_path = os.path.splitext(file_path)[0] + "_extracted.txt"
        with open(save_path, "w", encoding="utf-8") as f:
            f.write(extracted_text)
        
        messagebox.showinfo("Success", f"Text extracted and saved to:\n{save_path}")
        
    except Exception as e:
        messagebox.showerror("Error", f"Failed to extract text:\n{e}")

# GUI
root = tk.Tk()
root.title("EasyOCR Scanner")
root.geometry("300x150")

label = tk.Label(root, text="Click below to scan an image for text", pady=20)
label.pack()

scan_button = tk.Button(root, text="Scan Image", command=scan_image, padx=20, pady=10)
scan_button.pack()

root.mainloop()
```
